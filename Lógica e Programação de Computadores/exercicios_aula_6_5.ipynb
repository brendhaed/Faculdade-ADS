```python
#Exercício 1 –Construa um programa que escreve o fatorial dos 100 primeiros inteiro.
numero = 0
while numero <=99:
  fat = 1
  aux = 1
  while aux <= numero:
    fat = fat * aux
    aux = aux + 1
  print("O fatorial de",numero,"é",fat)
  numero = numero + 1
```
